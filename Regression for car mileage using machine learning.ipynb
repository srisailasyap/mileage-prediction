```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0231EN-SkillsNetwork/datasets/mpg.csv"
d = pd.read_csv(URL)
```

```python
d.shape
```

```python
d.plot.scatter(x = "Horsepower", y = "MPG")
```

```python
target = d["MPG"]
```

```python
features = d[["Horsepower","Weight"]]
```

```python
l = LinearRegression()
```

```python
l.fit(features,target)
```

```python
l.score(features,target)
```

```python
l.predict([[100,2000]])
```
